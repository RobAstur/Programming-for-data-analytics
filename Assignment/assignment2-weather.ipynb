```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
# Used the pd.read.csv to open the file. Other option is to download the file in the laptop and open it.
weather = pd.read_csv("https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv")
```

```python
#Get the column values and using matplotlib to create the histogram

dryBulbTempe = weather["dryBulbTemperature_Celsius"]

plt.hist(dryBulbTempe, color="blue",edgecolor="magenta",  linestyle='--', alpha=0.5)
plt.title("dryBulbTemperature_Celsius")
plt.show() 
```
