```python
# Calculate porcentage change
price_property_index['Pct_Change'] = price_property_index['VALUE'].pct_change() * 100

# Calculate cumulative percentage change over time.
price_property_index['Cumulative_Pct_Change'] = (1 + price_property_index['Pct_Change'] / 100).cumprod() - 1

# convert cummulative change in %
price_property_index['Cumulative_Pct_Change'] = price_property_index['Cumulative_Pct_Change'] * 100

# this step is important to preven a NAN in the first column
df = price_property_index.dropna(subset=['Cumulative_Pct_Change'])

# Size of the plot
plt.figure(figsize=(12, 6))  

# Create the line plot with seaborne
sns.lineplot(data=df, x='Year', y='Cumulative_Pct_Change', marker='o', linewidth=2, color='b', label='Cumulative Change')

# Adding design the plot
# Adding grid
plt.grid(True, linestyle='--', alpha=0.7)

# Title and labels
plt.title('Accumulated Percentage Change in Property Value', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Accumulated Percentage Change (%)', fontsize=12)

# this helps to have full years in the x axe. Otherwise I am gettgin floats
plt.xticks(ticks=df['Year'], labels=[str(year) for year in df['Year']], rotation=45)

# Add a legend 
plt.legend(loc='upper left', fontsize=12)

# Tight layout to avoid overlap and ensure everything fits
plt.tight_layout()

# Show the plot
plt.show()
```
